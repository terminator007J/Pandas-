```
cars_data.insert(10,'PriceRange',"")
```

```
for i in range (0,len(cars_data['Price']),1):
   if (cars_data['Price'][i]<8450):
     cars_data['PriceRange'][i]='Low'
   elif (cars_data['Price'][i]>11500):
     cars_data['PriceRange'][i]='High'
   else: ['PriceRange'][i]='Medium'
```

```
cars_data.insert(11,'Ageconverted','')
```

```
cars_data.insert(12,'Kmpm','')
```

```
cars_data
```

```
def vc(val1,val2):
    permonth=val1/12
    ratio=val2/val1
    return(permonth,ratio)
```

```
cars_data['Ageconverted'],cars_data['Kmpm']=vc(cars_data['Age'],cars_data['KM'])
cars_data['Ageconverted']=round(cars_data['Ageconverted'])
cars_data['Kmpm']=round(cars_data['Kmpm'])
cars_data
```
